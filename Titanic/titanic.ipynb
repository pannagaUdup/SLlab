```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df=pd.read_csv('TitanicDataset.csv')
titanic_df.head()
```

```python
titanic_df['Survived']=titanic_df['Survived'].map({
    1:"survived",
    0:"died"
})
titanic_df.head(5)
```

```python
titanic_df.drop(['Parch','PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
titanic_df['Pclass']=titanic_df['Pclass'].map({
    1:"Luxury Class",
    2:"Economy Class",
    3:'Lower Class'
})
titanic_df.head(5)
```

```python
titanic_df ['Embarked'] = titanic_df ['Embarked'].map({
    'C':'Cherbourg',
    'Q':'Queenstown',
    'S':'Southampton'
})
titanic_df.head(5)
```

```python
ax=sns.countplot(x="Pclass", hue="Survived" ,palette="Set1",data=titanic_df)
ax.set(title="Passenger class vs Surrvived",xlabel="Passenger Class",ylabel="Survived")
plt.show()
```

```python
print(pd.crosstab(titanic_df["Sex"],titanic_df.Survived))
ax=sns.countplot(x="Sex",hue="Survived",palette="Set1",data=titanic_df)
ax.set(title='Gender vs survived',xlabel='Gender',ylabel='Survived')
plt.show()
```

```python
intervals=(0,18,35,60,120)
categories=["children","teen","adult","old"]
titanic_df["AgeCategories"]=pd.cut(titanic_df.Age,intervals,labels=categories)
ax=sns.countplot(x="AgeCategories",hue="Survived",palette="Set1",data=titanic_df)
ax.set(title="AgeCategories vs Survived",xlabel="AgeCategories",ylabel="Survived")
plt.show()
```
